## How to help ratings?

In this notebook we are trying to find some key recipe attributes which seem to result in higher ratings.

---

### Cooking healthy

Carbs are commonly associated with weight gain, while fiber is considered healthy by helping digestion with little or no body weight impact. Protein is an important ingredient of a healthy diet, while cholesterol is associated with cardiovascular diseases and diabetes. 

The plots below show that recipes with higher fiber/carb ratio tend to have higher ratings, as well as recipes with higher protein/cholesterol ratio. This suggests that posting recipes with **healthier nutritional decomposition** helps to gain **more favorable ratings**.

```python
plt.figure(figsize=(15,8))
plt.scatter(clean_df['dietaryfiber'] / clean_df['totalcarbs'], clean_df['rating'])

plt.title('Fiber vs Carbs correlated with Ratings', fontsize=20)
plt.xlabel('Fiber/Carbs ratio', fontsize=16)
plt.ylabel('Rating', fontsize=16)
```

```python
plt.figure(figsize=(15,8))
plt.scatter(clean_df['protein'] / clean_df['cholesterol'], clean_df['rating'])

plt.title('Protein vs Cholesterol correlated with Ratings', fontsize=20)
plt.xlabel('Protein/Cholesterol ratio', fontsize=16)
plt.ylabel('Rating', fontsize=16)
```

---

### Cook for more

Preparing a dish with more servings means that either it can be served for more people, or it can be served for the same amound of people over a longer period. In both cases, higher servings value mean less cooking over a given period of time and number of people.

The plot below shows that recipes to prepare dishes for **more servings** receive **higher rating** scores. 

```python
plt.figure(figsize=(15,8))
plt.scatter(clean_df['servings'], clean_df['rating'])

plt.title('Servings correlated with Ratings', fontsize=20)
plt.xlabel('Servings', fontsize=16)
plt.ylabel('Rating', fontsize=16)
```
