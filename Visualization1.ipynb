```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_json('./pickles/dataframes/all_recipe_cleaned.json')
```

# Visualization of key recipe attributes

### Cooking time
Some recipes contain no cooking time, and there are a few outliers with long cooking times. In order to see the distribution of most common cooking times, we visualized the number of recipes where cooking time is between zero and 120 minutes.

The results how that there are many recipes where cooking time is under 15 minutes, and most recipes fall into the category of cooking time of 1 hour or less.

```python
# focus on cooking time between 0 and 120 minutes
cook_time = data[(data['cook_time'] > 0) & (data['cook_time'] < 120)]
cook_time['cook_time'].plot(kind='hist', bins=24, figsize=(20,6), title='Cooking time distribution')
```

---

### Ratings
Ratings contain values between 0 and 2, recipes with rating of 2 contributing to 90% of all recipes.

```python
data['rating'].value_counts().plot(kind='barh', figsize=(20,4))
```

---

### Servings
Serving values range from 1 to 240, but recipes with servings above 50 contribute only about 1% of the total recipes, so we considered such values as outliers. Looking at the distribution of servings, the most typical values are between 4 and 12, with even numbers being much more common.

```python
servings = data[(data['servings'] > 0) & (data['servings'] < 50)]
servings['servings'].plot(kind='hist', bins=50, figsize=(20,6), title='Servings distribution')
```
